```python
query_country = '''
SELECT * 
  FROM Country;
'''

query_league = '''
SELECT * 
  FROM League;
'''

query_team = '''
SELECT * 
  FROM Team
'''

query_team_attributes = '''
SELECT * 
  FROM Team_Attributes;
'''

query_match = '''
SELECT * 
  FROM Match;
'''

query_player = '''
SELECT * 
  FROM Player;
'''

query_player_attributes = '''
SELECT * 
  FROM Player_Attributes;
'''
```
